```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv('GfSD\SDGData.csv', sep=',',decimal='.', skiprows=0)
df_list = []


df_  = df.iloc[[102347]] # Estados Unidos
df_list.append(df_)
df_ = df.iloc[[41147]] # Repblica Dominicana
df_list.append(df_)
df_ = df.iloc[[28907]] # Brasil
df_list.append(df_)

plt.style.use('default')
plt.figure()
for df in df_list:
    years = []
    values = []
    for year in range(2001,2021):
        year_ = str(year)
        dado_invalido = pd.isnull(df[year_])
        df.loc[dado_invalido, year_] = 0
        if year == 2010:
            df.loc[dado_invalido, year_] = (df[str(year-1)] + df[str(year+1)])/2
        if year >= 1994:
            years.append(year)
            values.append(df[year_])
    plt.plot(years,values)

plt.xticks(years, rotation=45)
plt.xlabel('Anos')
plt.ylabel('Porcentagem da populao do pas')
plt.title('Populao jovem sem educao, emprego ou treinamento, entre pases selecionados (%)')
plt.legend(['Estados Unidos','Repblica Dominicana', 'Brasil'])
plt.grid()
plt.show()
```

O objetivo escolhido foi o objetivo 4: Educao de Qualidade, o qual visa garantir educao de qualidade, inclusiva e equitativa, e promover oportunidades de aprendizado vitalcias para todos.  um objetivo importante de se mensurar, considerando a importncia da educao em todos os  processos cruciais da populao local. Educao de qualidade garante que indivduos saibam cumprir suas funes na sociedade com eficcia, trabalhem para que haja melhor desenvolvimento e indiretamente reduz os nveis de criminalidade.

Quanto ao marcador, escolhemos a porcentagem da populao jovem sem acesso a educao, emprego ou treinamento. A escolha foi feita baseada no somente na importncia de se mensurar tal parmetro para nosso propsito, considerando que  um bom indicador direto do nvel de educao, mas tambm por fornecer base para especulaes sobre o futuro da localidade.

A escolha do marcador faz sentido pelo fato de a porcentagem direta de acesso a educao explicar bem os nveis de educao de qualidade ou sua falta. Alm disso, fatorando emprego nos dados, isso garante mais eficcia, sabendo que ndices de emprego e educao se correlacionam rigorosamente. Por fim, ao filtrarmos aqueles que no esto recebendo treinamento, conseguimos definir uma parcela da populao vulnervel a crises de modo geral.

Analisando o grfico,  possvel perceber um aumento drstico no ano de 2020 em todos os pases. Isso se trata de uma consequncia direta da pandemia do vrus COVID-19, a qual gerou uma crise e causou uma queda em educao de qualidade mundialmente durante o perodo. Alm disso, percebe-se uma clara diferena entre os pases subdesenvolvidos e os Estados Unidos no quesito. Isso  de se esperar, considerando as diferenas socioeconmicas.

Quanto  relao entre o Brasil e a Repblica Dominicana,  possvel observar algo interessante: os dois possuem ndices parecidos entre quase todos os anos mensurados. No entanto, no ano de 2020, a Repblica Dominicana foi afetada mais extremamente que o Brasil. Isso demonstra que apesar de as duas naes serem to similares em ndices de educao, o Brasil possui uma capacidade muito melhor de preserv-los em caso de crise, o que se mostra certamente diferente na Repblica Dominicana.

Observao final: para o ano de 2010 no Brasil, havia falta de dados. Para evitar que o grfico fosse afetado visualmente, para tal ano foi feito data amputation, ou seja, os dados foram editados para propsitos de demonstrao.
